```python
def movie_title(link):
    titles=[]
    for li in link:
#         print(li)
        substring="//title/"
        if substring in l:
            response=requests.get(li)
            html =response.text
            soup=BeautifulSoup(html,'html.parser')
        #     print(soup)
            mtags=soup.select('title')
            mtitle=str(mtags)[8:]
            mtitle=mtitle.split("(")
            print(mtitle[0])
            titles.append(mtitle[0])
        else:
            titles.append('')
    return titles
#     title_tag="class=""
```

```python
def scrape(link):
    crew=[]
    cast=[]
    plot=[]
    review=[]
    genre=[]
    rating=[]
    data_of_released=[]

    for  l in link:

        response=requests.get(l)
        html =response.text
        soup=BeautifulSoup(html,'html.parser')
        soup=str(soup)
        sub="description"
        if sub in soup:
            s=soup.split(":")
            try:
    #             print(s[5].split("'data-id")[0].split(". ")[1])
                ct=s[5].split("'data-id")[0].split(". ")[1]
                cast.append(ct)
            except:
                cast.append(' ')
            try:
    #             print(s[13].split("review")[0])
                pt=s[13].split("review")[0]
                plot.append(pt)
            except:
                plot.append(' ')
            try:
    #             print(s[25])
                rw=s[25]
                review.append(rw)
            except:
                review.append(' ')
        else:
            cast.append(' ')
            plot.append(' ')
            review.append(' ')
            
        print("Done ",l)
    return cast,plot,review

#     print(crew)
#     print(cast)
#     print(plot)
#     print(review)
#     print(genre)
#     print(rating)
#     print(data_of_released)
    

```
